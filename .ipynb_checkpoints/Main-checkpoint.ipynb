```python
import cv2
from matplotlib import pyplot as plt
import numpy as np 
import random
import math
import numpy as np
import statistics 
import scipy.io
from scipy.stats import norm
mat = scipy.io.loadmat('a1digits.mat')


test_data = np.array(mat['digits_test']) 
training_data = np.array(mat['digits_train'])    

test_data.shape
```

```python
imgs = []
for i in range(0,10):
        temp = training_data[:,0,i] 
        temp = temp.reshape(8,8)
        imgs.append(temp)

counter = 0 
f, axarr = plt.subplots(1,10)
for img in imgs: 
    axarr[counter].imshow(img)
    counter +=1
```

```python
# fit a probability distribution to a univariate data sample
def fit_distribution(data):
    # estimate parameters
    mu = np.mean(data)
    sigma = np.std(data)
    #print(mu, sigma)
    # fit distribution
    dist = norm(mu, sigma)
    return dist
 

# sort data into classes
prior_dist = []

for i in range(0,9): 
    print(training_data.shape[1]*training_data.shape[2],training_data[:,:,i].shape[1])
    prior_dist.append(training_data[:,:,i].shape[1]/training_data.shape[1]*training_data.shape[2])
                  
print (prior_dist) 
```

```python
gaus_dist = [[] for i in range(10)]

for Class in range(0,10):
    for feature in range (0,64):
        gaus_dist[Class].append(fit_distribution(np.transpose(training_data[feature,:,Class])))

print(len(np.transpose(test_data[:,:,2])))
```

```python
result = []
count = 0
for cl1 in range(10):
    for testVector in np.transpose(test_data[:,:,cl1]):
        count = 0
        sample_probs = []
        for cl2 in range(10):
            sample_probs.append(np.prod([gaus_dist[cl2][i].pdf(testVector[i]) for i in range(64)])*10)
        sample_check = max(sample_probs)
        max_index = sample_probs.index(sample_check)
        
        result.append(sample_probs)
        count = count + 1
```

```python
for i in range(10):
    sample_check = max(result[i])
    max_index = result[i].index(sample_check)
    print("This sample image belongs to class: " + str(max_index + 1))
    
    temp = np.transpose(test_data[:,:,0])[0] 
    temp = temp.reshape(8,8)
    plt.imshow(temp)
    
f, axarr = plt.subplots(1,10)
for img in imgs: 
    axarr[counter].imshow(img)
    counter +=1

temp = np.transpose(test_data[:,:,0])[0] 
temp = temp.reshape(8,8)
plt.imshow(temp)
```
